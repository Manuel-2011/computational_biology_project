```python
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
data_df = pd.read_csv('./basic_stats_results.tsv', sep='\t')
data_df
```

```python
data_df.groupby('Group').size()
```

```python
p_values = [None]
q_values = [0]
for i in range(1, len(data_df)):
    otu1 = data_df.loc[i-1, 'OTU_ID']
    otu2 = data_df.loc[i, 'OTU_ID']
    group1 = data_df.loc[i-1, 'Group']
    group2 = data_df.loc[i, 'Group']
    p_value = None
    if otu1 == otu2 and 'INU' in group2 and group1 in group2:
        print(otu1, group1, group2)
        try:
            mean1 = data_df.iloc[i-1]['Mean']
            mean2 = data_df.iloc[i]['Mean']
            std1 = data_df.iloc[i-1]['SD']
            std2 = data_df.iloc[i]['SD']
            n1 = data_df.iloc[i-1]['N']
            n2 = data_df.iloc[i]['N']

            t = (mean1 - mean2) / ((std1**2/n1) + (std2**2/n2))**0.5
            v = (std1**2/n1 + std2**2/n2)**2 / ((std1**2/n1)**2/(n1-1) + (std2**2/n2)**2/(n2-1))
            v = round(v)
        

            p_value = 2 * (1 - stats.t.cdf(abs(t), df=v))
        except:
            print('Error in>', otu1, group1, group2)
    p_values.append(p_value)

data_df['P_value'] = p_values

# Benjamini-Hochberg correction
data_df = data_df.sort_values(by='P_value')
data_df['Q_value'] = data_df['P_value'] * np.count_nonzero(data_df['P_value']) / data_df.index
```
